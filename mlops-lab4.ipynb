```python
code = r'''
from __future__ import annotations

import os
import pickle
from dataclasses import dataclass
from tempfile import TemporaryDirectory
from typing import Tuple

import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

import mlflow
import mlflow.pyfunc
from mlflow.tracking import MlflowClient

import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
from mlflow.models.signature import infer_signature

FEATURES = ["pressure", "flow", "radius"]
TARGETS = ["energy", "power", "elastic_energy"]

def _make_sequences(X: np.ndarray, y: np.ndarray, seq_len: int) -> Tuple[np.ndarray, np.ndarray]:
    if seq_len < 1:
        raise ValueError("seq_len must be >= 1")
    if len(X) <= seq_len:
        raise ValueError(f"Not enough rows ({len(X)}) for seq_len={seq_len}")

    Xs, ys = [], []
    for i in range(len(X) - seq_len):
        Xs.append(X[i : i + seq_len])
        ys.append(y[i + seq_len])
    return np.asarray(Xs, dtype=np.float32), np.asarray(ys, dtype=np.float32)

class LSTMRegressor(nn.Module):
    def __init__(self, n_features: int, hidden_size: int, num_layers: int, dropout: float, n_outputs: int):
        super().__init__()
        self.lstm = nn.LSTM(
            input_size=n_features,
            hidden_size=hidden_size,
            num_layers=num_layers,
            batch_first=True,
            dropout=dropout if num_layers > 1 else 0.0,
        )
        self.head = nn.Sequential(
            nn.Linear(hidden_size, hidden_size),
            nn.ReLU(),
            nn.Linear(hidden_size, n_outputs),
        )

    def forward(self, x):
        out, _ = self.lstm(x)
        last = out[:, -1, :]
        return self.head(last)

@dataclass
class TrainCfg:
    seq_len: int = int(os.getenv("SEQ_LEN", "10"))
    hidden_size: int = int(os.getenv("HIDDEN_SIZE", "64"))
    num_layers: int = int(os.getenv("NUM_LAYERS", "2"))
    dropout: float = float(os.getenv("DROPOUT", "0.1"))
    batch_size: int = int(os.getenv("BATCH_SIZE", "64"))
    lr: float = float(os.getenv("LR", "1e-3"))
    epochs: int = int(os.getenv("EPOCHS", "25"))
    test_size: float = float(os.getenv("TEST_SIZE", "0.2"))
    random_state: int = int(os.getenv("RANDOM_STATE", "42"))

class EnergyLSTMPyFunc(mlflow.pyfunc.PythonModel):
    def load_context(self, context):
        with open(context.artifacts["x_scaler"], "rb") as f:
            self.x_scaler = pickle.load(f)
        with open(context.artifacts["y_scaler"], "rb") as f:
            self.y_scaler = pickle.load(f)

        ckpt = torch.load(context.artifacts["model_pt"], map_location="cpu")
        self.seq_len = int(ckpt["seq_len"])
        self.features = ckpt["features"]
        self.targets = ckpt["targets"]

        model = LSTMRegressor(
            n_features=len(self.features),
            hidden_size=int(ckpt["hidden_size"]),
            num_layers=int(ckpt["num_layers"]),
            dropout=float(ckpt["dropout"]),
            n_outputs=len(self.targets),
        )
        model.load_state_dict(ckpt["state_dict"])
        model.eval()
        self.model = model

    def predict(self, context, model_input: pd.DataFrame) -> pd.DataFrame:
        df = model_input.copy()
        missing = [c for c in self.features if c not in df.columns]
        if missing:
            raise ValueError(f"Missing required fields: {missing}")

        X = df[self.features].to_numpy(dtype=np.float32)
        Xs = self.x_scaler.transform(X).astype(np.float32)
        X_seq = Xs.reshape(len(Xs), 1, len(self.features))

        with torch.no_grad():
            y_pred_scaled = self.model(torch.from_numpy(X_seq)).numpy()

        y_pred = self.y_scaler.inverse_transform(y_pred_scaled)
        return pd.DataFrame(y_pred, columns=self.targets)

def train_baseline_from_pandas(df: pd.DataFrame, model_name: str, experiment: str) -> dict:
    cfg = TrainCfg()

    mlflow.set_experiment(experiment)

    # Ensure numeric
    for c in FEATURES + TARGETS:
        df[c] = pd.to_numeric(df[c], errors="coerce")
    df = df.dropna(subset=FEATURES + TARGETS).reset_index(drop=True)

    X = df[FEATURES].to_numpy(dtype=np.float32)
    y = df[TARGETS].to_numpy(dtype=np.float32)

    X_seq, y_seq = _make_sequences(X, y, seq_len=cfg.seq_len)

    Xtr, Xte, ytr, yte = train_test_split(
        X_seq, y_seq, test_size=cfg.test_size, random_state=cfg.random_state, shuffle=True
    )

    x_scaler = StandardScaler()
    y_scaler = StandardScaler()

    Xtr_flat = Xtr.reshape(-1, len(FEATURES))
    x_scaler.fit(Xtr_flat)
    y_scaler.fit(ytr)

    def scale_X(Xs: np.ndarray) -> np.ndarray:
        flat = Xs.reshape(-1, len(FEATURES))
        flat_s = x_scaler.transform(flat).astype(np.float32)
        return flat_s.reshape(Xs.shape[0], Xs.shape[1], Xs.shape[2])

    Xtr_s = scale_X(Xtr)
    Xte_s = scale_X(Xte)
    ytr_s = y_scaler.transform(ytr).astype(np.float32)
    yte_s = y_scaler.transform(yte).astype(np.float32)

    device = "cuda" if torch.cuda.is_available() else "cpu"

    model = LSTMRegressor(
        n_features=len(FEATURES),
        hidden_size=cfg.hidden_size,
        num_layers=cfg.num_layers,
        dropout=cfg.dropout,
        n_outputs=len(TARGETS),
    ).to(device)

    train_ds = TensorDataset(torch.from_numpy(Xtr_s), torch.from_numpy(ytr_s))
    test_ds = TensorDataset(torch.from_numpy(Xte_s), torch.from_numpy(yte_s))
    train_loader = DataLoader(train_ds, batch_size=cfg.batch_size, shuffle=True)
    test_loader = DataLoader(test_ds, batch_size=cfg.batch_size, shuffle=False)

    opt = torch.optim.Adam(model.parameters(), lr=cfg.lr)
    loss_fn = nn.MSELoss()

    with mlflow.start_run() as run:
        mlflow.log_param("model_type", "LSTMRegressor")
        mlflow.log_param("seq_len", cfg.seq_len)
        mlflow.log_param("hidden_size", cfg.hidden_size)
        mlflow.log_param("num_layers", cfg.num_layers)
        mlflow.log_param("dropout", cfg.dropout)
        mlflow.log_param("batch_size", cfg.batch_size)
        mlflow.log_param("lr", cfg.lr)
        mlflow.log_param("epochs", cfg.epochs)
        mlflow.log_param("features", ",".join(FEATURES))
        mlflow.log_param("targets", ",".join(TARGETS))
        mlflow.log_param("device", device)

        for epoch in range(1, cfg.epochs + 1):
            model.train()
            train_losses = []
            for xb, yb in train_loader:
                xb = xb.to(device)
                yb = yb.to(device)
                opt.zero_grad()
                pred = model(xb)
                loss = loss_fn(pred, yb)
                loss.backward()
                opt.step()
                train_losses.append(loss.item())

            model.eval()
            val_losses = []
            with torch.no_grad():
                for xb, yb in test_loader:
                    xb = xb.to(device)
                    yb = yb.to(device)
                    pred = model(xb)
                    val_losses.append(loss_fn(pred, yb).item())

            mlflow.log_metric("train_mse", float(np.mean(train_losses)), step=epoch)
            mlflow.log_metric("val_mse", float(np.mean(val_losses)), step=epoch)

        model.eval()
        with torch.no_grad():
            y_pred_scaled = model(torch.from_numpy(Xte_s).to(device)).cpu().numpy()

        y_pred = y_scaler.inverse_transform(y_pred_scaled)
        y_true = yte

        rmse = float(np.sqrt(mean_squared_error(y_true, y_pred)))
        mlflow.log_metric("rmse", rmse)

        with TemporaryDirectory() as tmpdir:
            x_scaler_path = os.path.join(tmpdir, "x_scaler.pkl")
            y_scaler_path = os.path.join(tmpdir, "y_scaler.pkl")
            model_path = os.path.join(tmpdir, "model.pt")

            with open(x_scaler_path, "wb") as f:
                pickle.dump(x_scaler, f)
            with open(y_scaler_path, "wb") as f:
                pickle.dump(y_scaler, f)

            torch.save(
                {
                    "state_dict": model.cpu().state_dict(),
                    "seq_len": cfg.seq_len,
                    "hidden_size": cfg.hidden_size,
                    "num_layers": cfg.num_layers,
                    "dropout": cfg.dropout,
                    "features": FEATURES,
                    "targets": TARGETS,
                },
                model_path,
            )

            # Important: include THIS FILE so serving can import LSTMRegressor etc.
            input_example = df[FEATURES].head(5).copy()

            # Make a matching output_example by running the *inference path* (seq_len=1)
            model_cpu = model.cpu()
            model_cpu.eval()

            X_ex = input_example[FEATURES].to_numpy(dtype=np.float32)
            X_ex_s = x_scaler.transform(X_ex).astype(np.float32)
            X_ex_seq = X_ex_s.reshape(len(X_ex_s), 1, len(FEATURES))

            with torch.no_grad():
                y_ex_scaled = model_cpu(torch.from_numpy(X_ex_seq)).numpy()

            y_ex = y_scaler.inverse_transform(y_ex_scaled)
            output_example = pd.DataFrame(y_ex, columns=TARGETS)

            signature = infer_signature(input_example, output_example)

            # --- Log model (use code_paths, not code_path) ---
            mlflow.pyfunc.log_model(
                artifact_path="model",
                python_model=EnergyLSTMPyFunc(),
                artifacts={
                    "model_pt": model_path,
                    "x_scaler": x_scaler_path,
                    "y_scaler": y_scaler_path,
                },
                registered_model_name=model_name,
                code_paths=[__file__],
                input_example=input_example,
                signature=signature,
            )

        client = MlflowClient()
        versions = client.search_model_versions(f"name='{model_name}'")
        version = None
        for v in sorted(versions, key=lambda x: int(x.version), reverse=True):
            if v.run_id == run.info.run_id:
                version = v.version
                break

        return {"run_id": run.info.run_id, "rmse": rmse, "model_name": model_name, "model_version": version}
'''
dbutils.fs.put(local_train_core_path , code, overwrite=True)
print("Wrote:", local_train_core_path)
```

```python
from mlflow.tracking import MlflowClient

client = MlflowClient()

new_version = str(result["model_version"])
new_rmse = float(result["rmse"])

PROD_ALIAS = "prod"
CANDIDATE_ALIAS = "staging"   # or "candidate" if you prefer

def try_get_alias_version(model_name: str, alias: str):
    try:
        mv = client.get_model_version_by_alias(model_name, alias)
        return str(mv.version), mv.run_id
    except Exception:
        return None, None

prod_version, prod_run_id = try_get_alias_version(MODEL_NAME, PROD_ALIAS)

if prod_version is None:
    print(f"No @{PROD_ALIAS} yet -> set @{PROD_ALIAS} to v{new_version}")
    client.set_registered_model_alias(MODEL_NAME, PROD_ALIAS, new_version)
else:
    prod_run = client.get_run(prod_run_id)
    prod_rmse = float(prod_run.data.metrics.get("rmse", float("inf")))

    print(f"Current @{PROD_ALIAS}: v{prod_version} rmse={prod_rmse}")
    print(f"New candidate:        v{new_version} rmse={new_rmse}")

    if new_rmse < (prod_rmse - PROMOTE_THRESHOLD):
        print(f" Better -> move @{PROD_ALIAS} to v{new_version}")
        client.set_registered_model_alias(MODEL_NAME, PROD_ALIAS, new_version)
    else:
        print(f" Not better -> set @{CANDIDATE_ALIAS} to v{new_version}")
        client.set_registered_model_alias(MODEL_NAME, CANDIDATE_ALIAS, new_version)
```

```python
from databricks.sdk import WorkspaceClient
from databricks.sdk.service.serving import EndpointCoreConfigInput, ServedEntityInput, TrafficConfig, Route

w = WorkspaceClient()

ENDPOINT_NAME = "energy-prod"

served_entity_name = f"{MODEL_NAME.replace('.', '-')}-{prod_version}"  # internal name for routing

config = EndpointCoreConfigInput(
    served_entities=[
        ServedEntityInput(
            name=served_entity_name,
            entity_name=MODEL_NAME,         # UC model full name: catalog.schema.model
            entity_version=prod_version,    # version string
            workload_size="Small",
            scale_to_zero_enabled=True,
        )
    ],
    traffic_config=TrafficConfig(
        routes=[Route(served_model_name=served_entity_name, traffic_percentage=100)]
    ),
)

# Create if missing, else update config
try:
    w.serving_endpoints.get(ENDPOINT_NAME)
    w.serving_endpoints.update_config(name=ENDPOINT_NAME, served_entities=config.served_entities, traffic_config=config.traffic_config)
    print("Updated endpoint:", ENDPOINT_NAME)
except Exception:
    w.serving_endpoints.create(name=ENDPOINT_NAME, config=config)
    print("Created endpoint:", ENDPOINT_NAME)
```
